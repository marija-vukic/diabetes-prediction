## **Data Preprocessing**

- In our data, there is one particular feature which is one of the few continuous features that may have outliers, BMI.

```python
# visually look for outliers through a boxplot
plt.figure(figsize=(7, 5))
sns.boxplot(x=df['BMI'])
plt.title('BMI Boxplot')
plt.show()
```

```python
# looking for outliers through the 1.5 * IQR method
bmi_values = df.get('BMI')
quarter1 = np.percentile(bmi_values, 25)
quarter3 = np.percentile(bmi_values, 75)

IQR = quarter3-quarter1
lower = quarter1 - (1.5*IQR)
upper = quarter3 + (1.5*IQR)

df_no_outliers = df[(df.get('BMI') >= lower) & (df.get('BMI') <= upper)]
```

```python
from sklearn.ensemble import IsolationForest

# using Isolation Forest to detect and predict anomalies
isolationForest = IsolationForest()
isolationForest.fit(df_no_outliers)

predictions = isolationForest.predict(df_no_outliers)
cleaned_df = df_no_outliers.assign(Anomalies=predictions)
```

```python
# removing anomalies and placing it into a cleaned dataframe
cleaned_df = cleaned_df[cleaned_df.get('Anomalies') == 1]
cleaned_df = cleaned_df.drop(columns = 'Anomalies')
```

```python
# isolating target variable which is Diabetes
x = cleaned_df.drop('Diabetes', axis=1)
y = cleaned_df.get('Diabetes')
```

- As seen from the diabetes analysis, there is a major class imbalance between No diabetes, prediabetes, and diabetes. So, I will address this by oversampling the minority classes and create synthetic data to balance the data out.

```python
from imblearn.over_sampling import SMOTE

# fixing imbalances through imblearn SMOTE to oversample the minority classes
smote = SMOTE(sampling_strategy='auto')
x_resampled, y_resampled = smote.fit_resample(x, y)
```

```python
# new distribution
y_resampled.value_counts()
```

## **Data Splitting (Train/Test)**

```python
# creating training and testing datasets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_resampled, y_resampled, test_size=0.2)
```

## **Modeling: Decision Trees**

```python
from sklearn.tree import DecisionTreeClassifier

# creating and fitting Decision Trees for my classification algorithm
decisionTree = DecisionTreeClassifier()

decisionTree.fit(X_train, y_train)
```

```python
from sklearn import metrics

# creating predictions for training and testing data
training_v1 = decisionTree.predict(X_train)
pred_v1 = decisionTree.predict(X_test)

# getting accuracies for training and testing
print("Training Accuracy:", round(metrics.accuracy_score(y_train, training_v1), 4))
print("Testing Accuracy:", round(metrics.accuracy_score(y_test, pred_v1), 4))
```

- There seems to be signs of overfitting so I will perform hyperparameter tuning to address this issue

```python
from sklearn.model_selection import cross_val_score

# cross validation scores along with mean value
cv_scores = cross_val_score(decisionTree, x_resampled, y_resampled, cv=10)
print('CV Scores:', cv_scores)
print('Mean Cross Validation:', round(np.mean(cv_scores), 4))
```

```python
from sklearn.metrics import classification_report

# classification report for the model
report = classification_report(y_test, pred_v1, zero_division=0)
print(report)
```

## **Hyperparameter Tuning**

```python
# performing hyperparameter tuning through Randomized Search
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint, uniform

decisionTree = DecisionTreeClassifier()

param_dist = {
    'criterion': ['gini', 'entropy'],
    'max_depth': randint(1, 100),  # Example: A wide range from 1 to 100
    'min_samples_split': randint(2, 20),
    'min_samples_leaf': randint(1, 10),
    'max_features': ['sqrt', 'log2', None]
}

# creating randomized search object
random_search = RandomizedSearchCV(decisionTree, param_dist, n_iter=50, cv=10)

# fitting randomized search model with the data
random_search.fit(X_train,y_train)
```

```python
# get tuned model
tuned_model = random_search.best_estimator_

# create predictions
training_v2 = tuned_model.predict(X_train)
pred_v2 = tuned_model.predict(X_test)

# get accuracy on new model
print("Training Accuracy:", round(metrics.accuracy_score(y_train, training_v2), 4))
print("Testing Accuracy:", round(metrics.accuracy_score(y_test, pred_v2), 4))
```

```python
# cross validation scores along with mean value
cv_scores = cross_val_score(tuned_model, x_resampled, y_resampled, cv=10)
print('CV Scores:', cv_scores)
print('Mean Cross Validation:', round(np.mean(cv_scores), 4))
```

## **Model Insights**

```python
# find best parameters for that model
parameters = random_search.best_params_
print('Best parameters:', parameters)
```

```python
# display feature importances
important_features = {}
columns = df.columns[1:]

for i in range(len(columns)):
  feature = tuned_model.feature_importances_[i]
  if feature >= 0.1:
    important_features[columns[i]] = feature

sort_dct = {k: v for k, v in sorted(important_features.items(), key=lambda item: item[1], reverse=True)}

colors = ['royalblue', 'firebrick', 'seagreen', 'darkorange']
plt.bar(sort_dct.keys(), sort_dct.values(), color=colors);
```

```python
# classification report for the tuned model
report = classification_report(y_test, pred_v2, zero_division=0)
print(report)
```
